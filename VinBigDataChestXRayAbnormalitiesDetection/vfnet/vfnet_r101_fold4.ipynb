# MMDet Framework

MMDetection is an open source object detection toolbox based on PyTorch. It is a part of the OpenMMLab project developed by Multimedia Laboratory, CUHK.

# Features

1. **Modular Design**
    1. The detection framework is decomposed into different components. This gives the flexiblity to construct a customized object detection framework using different backbones and models.
    
    2. The framework mainly contains following parts:
    
        1. Config: This is the place where you get to set the configurations for the framwork like data dirs, num of epochs, gpus to use etc.
        2. mmdet: This module contains the files related to backbones, necks, heads and losses etc.
        
        3. Tools: This is the directory that contains utilities for training, testing and computing the evaluation metric.

```python
from mmcv import Config
# cfg = Config.fromfile('./configs/faster_rcnn/faster_rcnn_r50_caffe_fpn_mstrain_1x_coco.py')
cfg = Config.fromfile('./configs/vfnet/vfnet_r101_fpn_mdconv_c3-c5_mstrain_2x_coco.py')

```

```python
print(f'Config:\n{cfg.pretty_text}')
```

# Configuration Settings On BaseConfig

```python
!pwd
```

```python
from mmdet.apis import set_random_seed

# dataset_dir = '../vinbigdata-cocodataset/vinbigdata-coco-dataset-3xdownsampled'
dataset_dir = '../vinbigdata-cocodataset/vinbigdata-coco-dataset-with-wbf-3xdownsampled_fold0'

cfg.dataset_type = 'CocoDataset'
cfg.classes = ("Aortic_enlargement", "Atelectasis", "Calcification", "Cardiomegaly", "Consolidation", "ILD", "Infiltration", "Lung_Opacity", "Nodule/Mass", "Other_lesion", "Pleural_effusion", "Pleural_thickening", "Pneumothorax", "Pulmonary_fibrosis")

cfg.data.train.img_prefix = f'{dataset_dir}/'
cfg.data.train.classes = cfg.classes
cfg.data.train.ann_file = f'{dataset_dir}/train_annotations.json'
cfg.data.train.type='CocoDataset'


cfg.data.val.img_prefix = f'{dataset_dir}/'
cfg.data.val.classes = cfg.classes
cfg.data.val.ann_file = f'{dataset_dir}/val_annotations.json'
cfg.data.val.type='CocoDataset'



cfg.data.test.img_prefix = f'{dataset_dir}/'
cfg.data.test.classes = cfg.classes
cfg.data.test.ann_file = f'{dataset_dir}/val_annotations.json'
cfg.data.test.type='CocoDataset'
cfg.data.samples_per_gpu = 4

# Train pipeline
cfg.train_pipeline = [
    dict(type='LoadImageFromFile'),
    dict(type='LoadAnnotations', with_bbox=True),
    dict(type='Resize', img_scale=(1333, 800), keep_ratio=True),
    dict(type='RandomFlip', flip_ratio=0.5),
    # dict(type="Shear", level=0),
    dict(type="Rotate", level=0, max_rotate_angle=10),
    dict(
        type='Normalize',
        mean=[123.675, 116.28, 103.53],
        std=[58.395, 57.12, 57.375],
        to_rgb=True),
    dict(type='Pad', size_divisor=32),
    dict(type='DefaultFormatBundle'),
    dict(type='Collect', keys=['img', 'gt_bboxes', 'gt_labels'])
]

cfg.model.bbox_head.num_classes = 14

cfg.data.train.type = 'CocoDataset'
cfg.data.val.type = 'CocoDataset'
cfg.data.test.type = 'CocoDataset'

cfg.optimizer.lr = 0.01
# cfg.lr_config.warmup = None
cfg.log_config.interval = 8

# Change the evaluation metric since we use customized dataset.
cfg.evaluation.metric = 'bbox'
# We can set the evaluation interval to reduce the evaluation times
cfg.evaluation.interval = 8
# We can set the checkpoint saving interval to reduce the storage cost
cfg.checkpoint_config.interval = 8

# Set seed thus the results are more reproducible
cfg.seed = 42
set_random_seed(42, deterministic=False)
cfg.gpu_ids = range(1)

# we can use here mask_rcnn.
cfg.load_from = './mmdetection_checkpoints/vfnet_r101_fpn_mdconv_c3-c5_mstrain_2x_coco_20201027pth-7729adb5.pth'
cfg.work_dir = "../mmdetection_vinbigdata_output/vfnet_r101_fold0_v1"

# One Epoch takes around 18 mins
cfg.runner.max_epochs = 16
# cfg.total_epochs = 24
```

```python
print(f'Config:\n{cfg.pretty_text}')
```

```python
# the dataset has been taken : https://www.kaggle.com/sreevishnudamodaran/vinbigdata-fusing-bboxes-coco-dataset
import os
dataset_dir = '../vinbigdata-cocodataset/vinbigdata-coco-dataset-with-wbf-3xdownsampled_fold0'
os.listdir(f"{dataset_dir}")
```

```python
# # for saving checkpoint and plots
import os
os.makedirs('../mmdetection_vinbigdata_output/vfnet_r101_fold0_v1')
```

```python
from mmdet.datasets import build_dataset
from mmdet.models import build_detector
from mmdet.apis import train_detector
```

# Initializing Detector

```python
model = build_detector(
        cfg.model,
        train_cfg=cfg.get('train_cfg'),
        test_cfg=cfg.get('test_cfg'))
```

# Building Dataset

```python
datasets = [build_dataset(cfg.data.train)]
```

# Training Detector With BBox Eval

```python
# cfg.lr_config.policy='step'
train_detector(model, datasets, cfg, distributed=False, validate=True)
```
